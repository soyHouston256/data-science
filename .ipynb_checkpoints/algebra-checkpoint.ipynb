```python
def space(v1,v2,limit):
    for a in range(-10,10):
        for b in range(-10,10):
            plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                        marker='.',
                        color="orange")
    plt.xlim(-limit,limit)
    plt.ylim(-limit,limit)
    
    plt.axvline(x=0, color='gray')
    plt.axhline(y=0, color='gray')
    
    plt.show()

space(v1,v2,100)
```

```python
v1 = np.array([1,1])
v2 = np.array([-1,-1])
```

```python
space(v1,v2,25)
```

```python
v1 = np.array([1,0])
v2 = np.array([2,-3])
space(v1,v2,50)
```
