```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
#Los Markets disponibles son:
'''
Format Strings
character	description
'.'	point marker
','	pixel marker
'o'	circle marker
'v'	triangle_down marker
'^'	triangle_up marker
'<'	triangle_left marker
'>'	triangle_right marker
'1'	tri_down marker
'2'	tri_up marker
'3'	tri_left marker
'4'	tri_right marker
'8'	octagon marker
's'	square marker
'p'	pentagon marker
'P'	plus (filled) marker
'*'	star marker
'h'	hexagon1 marker
'H'	hexagon2 marker
'+'	plus marker
'x'	x marker
'X'	x (filled) marker
'D'	diamond marker
'd'	thin_diamond marker
'|'	vline marker
'_'	hline marker

Line Styles
character	description
'-'	solid line style
'--'	dashed line style
'-.'	dash-dot line style
':'	dotted line style

Colors
character	color
'b'	blue
'g'	green
'r'	red
'c'	cyan
'm'	magenta
'y'	yellow
'k'	black
'w'	white'''
```

```python
plt.plot(x,y, 'cx-')
plt.show()
```

```python
plt.plot(x,y, 'cx:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'cD:')
plt.subplot(1,3,2)
plt.hist(y)
plt.subplot(1,3,3)
plt.pie(x)
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,1])
axes2 = fig.add_axes([0.1,0.55,0.2,0.3])

axes.plot(x,y,'b')

axes2.plot(y,x,'r')
axes2.set_facecolor('grey')
fig.show()
```

```python
y= np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (axes1, axes2) = plt.subplots(nrows=1, ncols=2)
#es como la desestructuracion de JS
axes1.plot(x,y,'b')
axes2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0,0].plot(x,y,'b')
axes[0,1].plot(y,x,'r')
axes[0,2].hist(x)
axes[0,3].pie(x)
axes[1,0].plot(x,np.sin(x),'b')
axes[1,1].plot(x,np.tan(x),'r')
axes[1,2].hist(x)
axes[1,3].plot(x, 1/np.tan(x))
```
