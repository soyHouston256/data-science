```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
sns.set(style='dark', palette='dark', font='Verdana', font_scale=1)
sns.barplot(x=['A','B','C'], y=[1,3,2], hue=['A','B','C'])
plt.show()
```

```python
tips =sns.load_dataset('tips')
```

```python
sns.displot(data=tips,x='total_bill')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip', hue='sex', kind='kde', legend=True, palette='Set2', alpha=0.5)
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative = False)
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative = False, hue='sex', stat='probability')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative = False, hue='sex', stat='percent')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative = False, hue='sex', stat='density')
plt.show()
```

```python
#stat for default is count
sns.histplot(data=tips, x='tip', bins=15, cumulative = False, hue='sex', stat='count')
plt.show()
```

```python
# multiple='layer', 'dodge', fill
sns.histplot(data=tips, x='tip', bins=15, cumulative = False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative= False, fill=True, bw_adjust=1)
plt.savefig('name.png')
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
#find = kde, hist
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple='stack')
plt.show()
```

```python
mean = tips.groupby(['sex', 'day'], observed=True)['tip'].mean()
mean
```

```python
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.title('Promedio de Propinas por DÃ­a y Sexo')
plt.xlabel('DÃ­a')
plt.ylabel('Propina Promedio ($)')
plt.show()
```

```python
sns.catplot(data=tips, 
            x='day', 
            y='tip', 
            hue='sex', 
            kind='bar',
            height=6, 
            aspect=1.5)
plt.title('Promedio de Propinas por DÃ­a y Sexo')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='tip', hue='sex')
plt.title('DistribuciÃ³n de Propinas por DÃ­a y Sexo')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x='day', y='tip', hue='sex', split=True)
plt.title('DistribuciÃ³n de Propinas por DÃ­a y Sexo')
plt.show()
```

```python
markers = {'Lunch':'D','Dinner':'s'} 
sns.scatterplot(data=tips,hue='day', x='total_bill', y='tip', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show() 
```

```python
markers = {'Lunch':'D','Dinner':'s'} 
sns.relplot(data=tips,hue='day', x='total_bill', y='tip', style='time', size='size', markers=markers, col='time', legend='auto')
plt.show() 
```
